```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read your dataset
df = pd.read_csv("ecommerce_returns_synthetic_data.csv")

# Preprocessing
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Return_Date'] = pd.to_datetime(df['Return_Date'])
df['Is_Returned'] = df['Return_Status'].apply(lambda x: 1 if x == 'Returned' else 0)

# Set seaborn style
sns.set(style="whitegrid")

# 1. Return Rate by Product Category
plt.figure(figsize=(10, 6))
category_return = df.groupby('Product_Category')['Is_Returned'].mean().sort_values(ascending=False)
sns.barplot(x=category_return.values, y=category_return.index, palette="coolwarm")
plt.title("Return Rate by Product Category")
plt.xlabel("Return Rate")
plt.ylabel("Product Category")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the dataset (if not already loaded)
df = pd.read_csv("ecommerce_returns_synthetic_data.csv")

# Clean and feature engineer
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Return_Date'] = pd.to_datetime(df['Return_Date'])
df['Is_Returned'] = df['Return_Status'].apply(lambda x: 1 if x == 'Returned' else 0)

# Save to CSV
df.to_csv("ecommerce_returns_cleaned.csv", index=False)

```

```python
plt.figure(figsize=(8, 5))
shipping_return = df.groupby('Shipping_Method')['Is_Returned'].mean().sort_values(ascending=False)
sns.barplot(x=shipping_return.index, y=shipping_return.values, palette="viridis")
plt.title("Return Rate by Shipping Method")
plt.ylabel("Return Rate")
plt.xlabel("Shipping Method")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
reason_counts = df['Return_Reason'].value_counts()
sns.barplot(x=reason_counts.values, y=reason_counts.index, palette="magma")
plt.title("Distribution of Return Reasons")
plt.xlabel("Number of Returns")
plt.ylabel("Return Reason")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_cities = df.groupby('User_Location')['Is_Returned'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_cities.values, y=top_cities.index, palette="cubehelix")
plt.title("Top 10 Cities by Return Rate")
plt.xlabel("Return Rate")
plt.ylabel("User Location")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Is_Returned', y='Discount_Applied', data=df, palette="Set2")
plt.title("Discount Applied vs Return Status")
plt.xlabel("Return Status (0 = Not Returned, 1 = Returned)")
plt.ylabel("Discount Applied")
plt.tight_layout()
plt.show()
```
